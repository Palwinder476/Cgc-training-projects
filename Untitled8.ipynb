```python
import kagglehub
import pandas as pd
import os

path = kagglehub.dataset_download("nehalbirla/vehicle-dataset-from-cardekho")

dataset_file_path = os.path.join(path, "Car details v3.csv")

df_cars = pd.read_csv(dataset_file_path)

display(df_cars.head())
```

```python
import matplotlib.pyplot as plt

fuel_counts = df_cars['fuel'].value_counts()
petrol_count = fuel_counts.get('Petrol', 0)
diesel_count = fuel_counts.get('Diesel', 0)

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.bar(['Petrol', 'Diesel'], [petrol_count, diesel_count], color=['skyblue', 'lightcoral'])
plt.title('Fuel Type Distribution')
plt.xlabel('Fuel Type')
plt.ylabel('Number of Cars')
plt.xticks(rotation=65)
plt.tight_layout()
plt.show()
```
